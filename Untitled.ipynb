```python
def json_load(filename):
    if os.path.exists(filename):
        return json.load(open("filename", "r"))
    else:
        raise ValueError(f"{filename} does not exist")
```

```python
def daft(url, search_name):
    daft_links = json_load("daft_links.txt")
    new_links = []
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    # get links and remove None values
    all_links = [link.get("href") for link in soup.find_all('a') if link.get("href")]
    # get relevant links
    links = [link for link in all_links  if "/for-rent" in link]
    for link in links:
        listing_id = link.split("/")[-1]
        if not daft_links[search_name].get(link):
            daft_links[search_name][link] = get_daft_details(link)
            new_links.append((link, get_daft_details(link)))
    json.dump(daft_links, open("daft_links.txt"), "w")
    return new_links
```

```python
def get_names(soup):
    names = []
    results = soup.find_all(class_ = "search_result")
    for res in results:
        x = res.find_all(class_ = "sresult_address")
        name = str(x[0]).split("</a>")[0].split(">")[-1][1:]
        names.append(name)
    return names
```

```python
def get_daftnames(daftsoup):
    daftresults = daftsoup.find_all(class_ = "TitleBlock__Address-sc-1avkvav-7 knPImU")
    daftnames = []
    for res in daftresults:
        x = str(res).split("address")[-1][2:-4]
        daftnames.append(x)
    return daftnames
```

```python
def daftcheck(daftsoup, name, url):
    old_names = set(pickle.load(open(name, "rb")))
    new_names = set(get_daftnames(daftsoup))
    if old_names == new_names:
        return False
    elif len(old_names) > len(new_names):
            pickle.dump(get_daftnames(daftsoup), open(name, "wb"))
    else:
        pickle.dump(get_daftnames(daftsoup), open(name, "wb"))
        for i in range(100):
            alert()
            new = new_names.difference(old_names)
            for propname in new:
                print(propname)
            print(name, url)
            time.sleep(10)
    
    return True
```

```python
def check(soup):
    old_names = set(pickle.load(open("names", "rb")))
    new_names = set(get_names(soup))
    if old_names == new_names:
        return False
    elif len(old_names) > len(new_names):
            pickle.dump(get_names(soup), open("names", "wb"))
    else:
        pickle.dump(get_names(soup), open("names", "wb"))
        for i in range(100):
            alert()
            new = new_names.difference(old_names)
            for name in new:
                print(name.replace("\n", ""))
            print(url)
            time.sleep(10)
        return True
```

```python
def run_forever(gap = 3):
    runs = 0
    while True:
        runs += 1
        page = requests.get(renturl)
        soup = BeautifulSoup(page.content, 'html.parser')
        check(soup)
            
        for name, url in daftnames:
            daftpage = requests.get(url)
            daftsoup = BeautifulSoup(daftpage.content, "html.parser")
            daftcheck(daftsoup, name, url)

        if runs % (30*60/gap) == 0 or runs == 1:
            print("Nothing in past 30 minutes")
        time.sleep(gap)
```
