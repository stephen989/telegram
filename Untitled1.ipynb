```python
def check(moves):
    if ("L" in moves) and ("R" in moves):
        return False
    if ("U" in moves) and ("D" in moves):
        return False
    return True
```

```python
def all_possible_moves(n, prev_moves = False):
    if n-1 == 1:
        all_moves = MOVES
    elif prev_moves == False:
        all_moves = all_possible_moves(n-1)
    else:
        all_moves = prev_moves
    for i in range(1):
        new_moves = []
        for l in MOVES:
            for comb in all_moves:
                comb += l
                if check(comb[-2:]):
                    new_moves.append(comb)
        all_moves = new_moves
            
    return all_moves
```

```python
def find_ending(start, moves):
    n = int(math.sqrt(len(start)))
    wx, wy = find_w("".join(start), n)
    wpos = wx + n * wy
    for move in moves:
        new_grid = start.copy()
        if move == "D":
            newx = wx
            newy = wy - 1
        elif move == "U":
            newx = wx
            newy = wy + 1
        elif move == "R":
            newx = wx - 1
            newy = wy
        else:
            newx = wx + 1
            newy = wy
        swap_ind = newx + newy * n
#         print(newx, newy)
        if (newx < 0) or (newx >= n) or (newy < 0) or (newy >= n):
            return False
        new_grid[wpos] = start[swap_ind]
        new_grid[swap_ind] = start[wpos]
        wx, wy = newx, newy
        start = new_grid.copy()
        wpos = wx + n * wy
    return new_grid
        
        
        
            
```

```python
def get_primes(n, start = 3):
    primes = [2]
    for x in range(start, n+1):
        is_prime = True
        for y in range(3, int(math.sqrt(x)+1), 2):
            if x % y == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(x)
    return primes
```
